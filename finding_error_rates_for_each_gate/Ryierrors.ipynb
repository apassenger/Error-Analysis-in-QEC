```python
from sympy.physics.quantum import TensorProduct
i = np.array([[1, 0], [0, 1]])
x = np.array([[0, 1], [1, 0]])
y = np.array([[0,-1j],[1j,0]])
z = np.array([[1, 0], [0,-1]])
```

```python
from itertools import product
def coefficients(kraus_list,q_number):
    paulis = list(product([i,x,y,z],repeat=q_number))
    spaulis = list(product(['i','x','y','z'],repeat=q_number))
    #print(spaulis)
    result = []
    for k_index,kraus in enumerate(kraus_list,start=0):
        for index, pauli in enumerate(paulis, start=0):
            #kraus=np.matmul(super_h.conj().T,kraus)
            a = 1/4*(np.trace(np.matmul(kraus,TensorProduct(pauli[0],pauli[1]))))
            #a = (np.trace(np.matmul(kraus,TensorProduct(pauli[0],pauli[1]))))
            result.append((a,spaulis[index],k_index))
    return result
```

```python
coef = coefficients(kraus,2)
coef
```
